```python
# bài 7
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))

for n in range(a, b + 1):
    if n < 2:
        continue

    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(n)
```
