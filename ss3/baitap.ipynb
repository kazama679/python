```python
# bài 1
def tinh_hieu(a, b):
    hieu = a - b
    print("Hiệu của hai số là:", hieu)

def tinh_tich(a, b):
    return a * b

a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))

tinh_hieu(a, b)

tich = tinh_tich(a, b)
print("Tích của hai số là:", tich)
```

```python
# bài 2
def check(a):
    if a < 2:
        return False
    if a == 2:
        return True
    for i in range(2, a):
        if a % i == 0:  
            return False
    return True
print(check(5))
```

```python
# bài 3
def factorial(n):
    if n <1: print("nhập lớn hơn 0")
    sum = n
    for i in range(1, n):
        sum*=i
    return sum    

print(factorial(5))
```

```python
# bài 4
from math import pi
def circle_area(radius):
    if radius <1: print("nhập lớn hơn 0")
    return radius * radius * pi

print(circle_area(5))
```

```python
def is_leap_year(year):
    if year % 400 == 0:
        print("nhuận")
    elif year %4==0: print("Nhuận")
    elif year %4==0 and year %100!=0:print("k nhuận")
    else: print("k nhuận")
is_leap_year(2024)
```

```python
# bài 6
def sum_digits(n):
    n = abs(n) 
    total = 0
    while n > 0:
        total += n % 10 
        n //= 10 
    return total

print(sum_digits(12345))
```

```python
# bài 7
def handleReverseTheNumber(n):
    sign = -1 if n < 0 else 1 
    n = abs(n)
    reversed_number = 0
    while n > 0:
        reversed_number = reversed_number * 10 + (n % 10)
        n //= 10
    return sign * reversed_number

print(handleReverseTheNumber(12345))
```

```python
# bài 8
def is_perfect_number(n):
    if n <= 1:
        return False
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total == n

num = int(input("Nhập một số nguyên: "))

if is_perfect_number(num):
    print(num, "là số hoàn hảo")
else:
    print(num, "không phải số hoàn hảo")
```
