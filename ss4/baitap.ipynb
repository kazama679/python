```python
# bài 8
products = {
    "A": 120,
    "B": 80,
    "C": 300,
    "D": 50
}

if "A" in products:
    print("Giá của A:", products["A"])
else:
    print("Sản phẩm A đã bị xóa")

sorted_products = dict(sorted(products.items(), key=lambda x: x[1]))
print("Sau khi sắp xếp:", sorted_products)

to_delete = [k for k, v in products.items() if v > 200]
for k in to_delete:
    products.pop(k)

print("Sau khi xóa sản phẩm > 200:", products)
```

```python
# bài tổng hợp
import random
products=[
    {
        "mã sản phẩm": 1,
        "tên sản phẩm": "Lê",
        "số lượng": 22,
        "giá": 12000,
    },
    {
        "mã sản phẩm": 2,
        "tên sản phẩm": "Xoài",
        "số lượng": 44,
        "giá": 12000,
    },
    {
        "mã sản phẩm": 3,
        "tên sản phẩm": "Đào",
        "số lượng": 33,
        "giá": 12000,
    },
]

while True:
    print("Menu Quản lý:\n"
        "1. Thêm sản phẩm mới.\n"
        "2. Hiển thị danh sách sản phẩm.\n"
        "3. Tìm kiếm sản phẩm theo mã sản phẩm.\n"
        "4. Cập nhật thông tin sản phẩm.\n"
        "5. Xóa sản phẩm theo mã sản phẩm.\n"
        "0. Thoát chương trình.\n"
        "Khác: Vui lòng nhập lại.")
    x = int(input("chọn số: "))
    if x == 1:
        newProducts = {
            "mã sản phẩm": random.randint(1, 100000),
            "tên sản phẩm": input("Nhập tên sp mới: "),
            "số lượng": int(input("Nhập sl sp mới: ")),
            "giá":int(input("Nhập tên giá mới: "))
        }
        products.append(newProducts)
    elif x == 2:
        print(f"{'Mã':<5} {'Tên sản phẩm':<10} {'Số lượng':<10} {'Giá':<10} \n")
        print("-" * 40)
        for p in products:
            print(f"{p['mã sản phẩm']:<5} {p['tên sản phẩm']:<10} {p['số lượng']:<10} {p['giá']:<10}")
    elif x == 3:
        id = int(input("Nhập id sản phẩm muốn tìm"))
        check=-1
        for i in range(0, len(products)): 
            print(products[i]["mã sản phẩm"])
            if products[i]["mã sản phẩm"] == id:
                check = i
        if check == -1: print("không tìm thấy")
        else: print(products[check])
    elif x == 4:
        id = int(input("Nhập id sản phẩm muốn sửa"))
        check=-1
        for i in range(0, len(products)): 
            print(products[i]["mã sản phẩm"])
            if products[i]["mã sản phẩm"] == id:
                check = i
        if check == -1: print("không tìm thấy")
        else: 
            editProducts = {
                "mã sản phẩm": products[check]["mã sản phẩm"],
                "tên sản phẩm": input("Nhập tên sp sửa: "),
                "số lượng": int(input("Nhập sl sp sửa: ")),
                "giá":int(input("Nhập tên giá sửa: "))
            }
            products[check] = editProducts
    elif x == 5:
        id = int(input("Nhập id sản phẩm muốn xóa"))
        check=-1
        for i in range(0, len(products)): 
            print(products[i]["mã sản phẩm"])
            if products[i]["mã sản phẩm"] == id:
                check = i
        if check == -1: print("không tìm thấy")
        else: 
            products.pop(check)
    elif x == 0:
        break
    else:
        print("lỗi")
```
