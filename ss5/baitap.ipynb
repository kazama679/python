```python
# bài 2
import csv

high_grade_students = []

try:
    with open("students.csv", "r", encoding="utf-8", newline="") as file:
        reader = csv.reader(file)
        try:
            header = next(reader)  
        except StopIteration:
            header = None

        for row in reader:
            if not row or len(row) < 3:
                continue
            name = row[0].strip()
            age = row[1].strip()
            grade_str = row[2].strip()
            try:
                grade = float(grade_str)
            except ValueError:
                continue  
            if grade >= 7:
                high_grade_students.append([name, age, grade])
except FileNotFoundError:
    print("Không tìm thấy file students.csv")
    high_grade_students = []

print("Danh sách học sinh có điểm >= 7:")
for student in high_grade_students:
    print(student)

with open("high_grade_students.csv", "w", encoding="utf-8", newline="") as out_file:
    writer = csv.writer(out_file)
    writer.writerow(["name", "age", "grade"])
    writer.writerows(high_grade_students)

print("Đã ghi dữ liệu vào high_grade_students.csv")
```

```python
# bài 3
import json

try:
    with open("user_data.json", "r", encoding="utf-8") as f:
        try:
            data = json.load(f)
        except json.JSONDecodeError:
            data = []
except FileNotFoundError:
    data = []

if isinstance(data, list):
    users = data
    container_type = "list"
elif isinstance(data, dict):
    if isinstance(data.get("users"), list):
        users = data["users"]
        container_type = "dict_with_users"
    else:
        vals = [v for v in data.values() if isinstance(v, dict) and "id" in v]
        if vals:
            users = vals
            container_type = "dict_values"
        else:
            users = []
            container_type = "dict_other"
else:
    users = []
    container_type = "other"

person = next((u for u in users if u.get("id") == 3), None)
if person:
    print(f"Người có id=3: name={person.get('name')}, age={person.get('age')}")
else:
    print("Không tìm thấy người có id = 3")

new_id = max([u.get("id", 0) for u in users] + [0]) + 1
new_user = {"id": new_id, "name": "Người mới", "age": 25}
users.append(new_user)

if container_type == "list":
    data_to_write = users
elif container_type == "dict_with_users":
    data["users"] = users
    data_to_write = data
else:
    data_to_write = {"users": users}

with open("user_data.json", "w", encoding="utf-8") as f:
    json.dump(data_to_write, f, ensure_ascii=False, indent=2)

print(f"Đã thêm người dùng mới với id={new_id} và ghi vào user_data.json")
```

```python
# bài 6
import json

try:
    with open("inventory.json", "r", encoding="utf-8") as f:
        try:
            data = json.load(f)
        except json.JSONDecodeError:
            data = []
except FileNotFoundError:
    data = []

if isinstance(data, dict) and isinstance(data.get("products"), list):
    products = data["products"]
    container_type = "dict_with_products"
elif isinstance(data, list):
    products = data
    container_type = "list"
else:
    products = []
    container_type = "other"

name = input("Tên sản phẩm mới: ").strip() or "Sản phẩm mới"
try:
    quantity = int(input("Số lượng: ").strip() or "0")
except ValueError:
    quantity = 0
try:
    price = float(input("Giá (ví dụ 19.99): ").strip() or "0")
except ValueError:
    price = 0.0

new_id = max([p.get("product_id", 0) for p in products] + [0]) + 1
new_product = {"product_id": new_id, "name": name, "quantity": quantity, "price": price}
products.append(new_product)

if container_type == "dict_with_products":
    data["products"] = products
    data_to_write = data
elif container_type == "list":
    data_to_write = products
else:
    data_to_write = {"products": products}

with open("inventory.json", "w", encoding="utf-8") as f:
    json.dump(data_to_write, f, ensure_ascii=False, indent=2)

print(f"Đã thêm sản phẩm mới với product_id={new_id}")
```

```python
# bài 8
from datetime import datetime, date

today = date.today()
try:
    with open("log.txt", "r", encoding="utf-8") as f:
        for line in f:
            line = line.rstrip("\n")
            if not line.strip():
                continue
            parts = line.split(maxsplit=2) 
            if len(parts) < 2:
                continue
            ts_str = parts[0] + " " + parts[1]
            ts = None
            try:
                ts = datetime.fromisoformat(ts_str)
            except ValueError:
                for fmt in ("%Y-%m-%d %H:%M:%S", "%Y/%m/%d %H:%M:%S", "%d-%m-%Y %H:%M:%S", "%d/%m/%Y %H:%M:%S"):
                    try:
                        ts = datetime.strptime(ts_str, fmt)
                        break
                    except ValueError:
                        pass
            if ts is None:
                continue
            if ts.date() == today:
                print(line)
except FileNotFoundError:
    print("Không tìm thấy file log.txt")
```
