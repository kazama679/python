```python
# bài 3
import numpy as np
np.random.seed(42)

A = np.random.randint(1, 100, size=(3, 3))
print(A)
print('\nMin:', A.min())
print('Max:', A.max())
print('Sum:', A.sum())
print('Mean:', A.mean())
print('STD:', A.std())
print('\nSumOfColumn:', A.sum(axis=0))
print('MeanOfColumn:', A.mean(axis=0))
print('\nSumOfRow:', A.sum(axis=1))
print('MeanOfRow:', A.mean(axis=1))
```

```python
# bài 4
import numpy as np
np.random.seed(0)

A = np.random.randint(1, 20, size=(3, 3))
B = np.array([10, 20, 30])     
B_row = B.reshape(1, 3)      
print('A:\n', A)
print('\nB:', B)

print('\nA + B (broadcast 1D):\n', A + B)
print('\nA - B_row (explicit 1x3):\n', A - B_row)
```

```python
# bài 5
import numpy as np
import time
np.random.seed(2)

a = np.random.randint(1, 21, size=10)
print('a =', a)

squares_vec = np.square(a)
roots_vec = np.sqrt(a)
print('\nsquares (vectorized):', squares_vec)
print('roots (vectorized):', roots_vec)

print('\nadd 5 (ufunc):', np.add(a, 5))
print('multiply by 2 (ufunc):', np.multiply(a, 2))

t0 = time.perf_counter()
s_loop = [x*x for x in a]
t1 = time.perf_counter()
t_loop = t1 - t0
t0 = time.perf_counter()
s_vec = a * a
t1 = time.perf_counter()
t_vec = t1 - t0
print(f'\nloop time: {t_loop:.6f} s, vectorized time: {t_vec:.6f} s')
```

```python
# bài 8
import numpy as np
import time
np.random.seed(5)

A = np.random.randint(1, 20, size=(5, 5))
B = np.array([10, 20, 30, 40, 50])
print('A:\n', A)
print('\nB:', B)

sum_b = A + B  
diff_b = A - B
sqrt_A = np.sqrt(A)
add5 = np.add(A, 5)
print('\nKết quả cộng (broadcasting):\n', sum_b)
print('\nKết quả trừ (broadcasting):\n', diff_b)
print('\nKết quả căn bậc hai (ufunc):\n', np.round(sqrt_A, 8))
print('\nKết quả cộng 5 vào mảng (ufunc):\n', add5)

t0 = time.perf_counter()
out_loop = np.empty_like(A)
for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        out_loop[i, j] = A[i, j] + B[j]
t1 = time.perf_counter()
t_loop = t1 - t0
t0 = time.perf_counter()
out_b = A + B
t1 = time.perf_counter()
t_vec = t1 - t0
print(f'\nThời gian vòng lặp: {t_loop:.6f} s, broadcasting: {t_vec:.6f} s')
```
