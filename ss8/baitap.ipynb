```python
# Bài 2:
from pathlib import Path
import json

base = Path('.') 
data_file = base / 'data.txt'
lines = [
    "Dòng 1: Đây là dòng một\n",
    "Dòng 2: Đây là dòng hai\n",
    "Dòng 3: Đây là dòng ba\n",
    "Dòng 4: Đây là dòng bốn\n",
    "Dòng 5: Đây là dòng năm\n",
]

with data_file.open('w', encoding='utf-8') as f:
    f.writelines(lines)

with data_file.open('r', encoding='utf-8') as f:
    read_lines = f.readlines()

for i, ln in enumerate(read_lines, start=1):
    print(f"Dòng {i}: {ln.rstrip()}")

student = {"name": "An", "age": 20, "scores": [8, 9, 7]}
student_file = base / 'student.json'
with student_file.open('w', encoding='utf-8') as f:
    json.dump(student, f, ensure_ascii=False, indent=2)

with student_file.open('r', encoding='utf-8') as f:
    student_from_file = json.load(f)

print("\nJSON đọc từ file:", student_from_file)
```

```python
# Bài 5
base = Path('.')
students_csv = base / 'students.csv'
scores_csv = base / 'scores.csv'
students_csv.write_text("1,An\n2,Binh\n3,Chau\n")
scores_csv.write_text("1,8\n2,9\n3,7\n")
df_students = pd.read_csv(students_csv, names=['id','name'])
df_scores = pd.read_csv(scores_csv, names=['id','score'])
merged = pd.merge(df_students, df_scores, on='id')
report = merged.groupby(['id','name'])['score'].agg(['min','max','mean']).reset_index()
report.columns = ['id','name','score_min','score_max','score_mean']
report.to_csv('report.csv', index=False)
print("Report:\n", report.to_csv(index=False))
```

```python

# Bài 6 
a = np.arange(1,5).reshape(4,1)
b = np.array([10,20,30,40,50]).reshape(1,5)
prod = a * b
sqrt_prod = np.sqrt(prod)
log_prod = np.log(prod)
stats_axis1 = {
    'min': prod.min(axis=1),
    'max': prod.max(axis=1),
    'mean': prod.mean(axis=1),
    'std': prod.std(axis=1)
}
print("a =\n", a)
print("b =\n", b)
print("a*b =\n", prod)
print("sqrt(a*b)=\n", np.round(sqrt_prod,4))
print("log(a*b)=\n", np.round(log_prod,4))
print("Stats axis=1:", stats_axis1, "\n")
```

```python

# Bài 7
data = {
    'name': ['An','Bình','Châu','Dũng'],
    'age': [20, np.nan, 22, 21],
    'score': [8.5, np.nan, 7.0, 9.0],
    'city': ['Hà Nội', 'HCM', np.nan, 'Hà Nội']
}
df7 = pd.DataFrame(data)
print("Original isna:\n", df7.isna(), "\n")
age_mean = df7['age'].mean()
score_median = df7['score'].median()
df7['age'] = df7['age'].fillna(age_mean)
df7['score'] = df7['score'].fillna(score_median)
df7['city'] = df7['city'].fillna('Unknown')
df7['city'] = df7['city'].astype('category')
def level(x):
    if x >= 8: return 'Good'
    if x >= 6: return 'Average'
    return 'Bad'
df7['score_level'] = df7['score'].apply(level)
# chuẩn hóa tuổi
age_mean2 = df7['age'].mean()
age_std2 = df7['age'].std(ddof=0)
df7['age_normalized'] = (df7['age'] - age_mean2) / age_std2
print("age_cleaned =", df7['age'].tolist())
print("score_cleaned =", df7['score'].tolist())
print("city (category) =", df7['city'])
print("score_level =", df7['score_level'].tolist())
print("age_normalized =", df7['age_normalized'].tolist())
```

```python
# bài 8
from pathlib import Path
import json
import pandas as pd
import re

logs_dir = Path('logs')
if not logs_dir.exists():
    print('Thư mục logs/ không tồn tại. Vui lòng tạo và thêm các file JSON.')
else:
    records = []
    date_re = re.compile(r"\d{4}-\d{2}-\d{2}")
    for p in sorted(logs_dir.glob('*.json')):
        m = date_re.search(p.stem)
        if not m:
            continue
        date_str = m.group(0)
        try:
            payload = json.loads(p.read_text(encoding='utf-8'))
        except Exception as e:
            print(f"Bỏ qua file không đọc được: {p.name} -> {e}")
            continue
        user = payload.get('user')
        action = payload.get('action')
        duration = payload.get('duration')
        try:
            duration = float(duration) if duration is not None else 0.0
        except Exception:
            print(f"duration không hợp lệ trong {p.name}, đặt về 0")
            duration = 0.0
        records.append({'date': pd.to_datetime(date_str).date(),
                        'user': user, 'action': action, 'duration': duration,
                        'source_file': p.name})

    if not records:
        print('Không tìm thấy record log hợp lệ trong logs/.')
    else:
        df_logs = pd.DataFrame.from_records(records)
        summary = df_logs.groupby('date', as_index=False)['duration'].sum()
        summary = summary.sort_values('date')
        for _, row in summary.iterrows():
            print(f"Date: {row['date']}, Total Duration: {row['duration']}")
        out_file = Path('daily_report.csv')
        summary.to_csv(out_file, index=False)
        print(f"Saved daily report to: {out_file.resolve()}")
```
